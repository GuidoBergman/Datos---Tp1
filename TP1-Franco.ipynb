```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df1 = pd.read_csv('train_values.csv')
```

```python
df1
```

```python
labels = pd.read_csv('train_labels.csv')
```

```python
labels
```

```python
df = df1.merge(labels, on='building_id')
```

```python
df
```

```python
df.age.value_counts().plot(kind='bar',figsize=(20,15))
```

```python
df.count_floors_pre_eq.value_counts().plot(kind='bar',figsize=(20,15))
```

```python
df.groupby('age').agg({'damage_grade':'mean'}).sort_values(ascending=False, by='damage_grade').head(10).plot(kind='bar',figsize=(15,10))
```

```python
df.groupby('height_percentage').agg({'damage_grade':'mean'}).sort_values(ascending=False, by='damage_grade').head(10).plot(kind='bar',figsize=(15,10))
```
